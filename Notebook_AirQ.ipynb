```python
def missing_values_table(df):
    """
    Generate a table of missing values in a dataset.

    Parameters:
    - df (pd.DataFrame): Input dataframe.
    - placeholder (scalar, optional): Value to be replaced with NaN.
    - sort_descending (bool): Sort the output by the percentage of missing values in descending order.
    - round_decimals (int): Number of decimals to round percentages.

    Returns:
    - pd.DataFrame: A DataFrame with columns for missing value counts and percentages.
    """
    # Replace placeholder with NaN with value -200
    df_with_placeholder = df.replace(-200, np.nan)

    # Part to calculate missing values and their percentages
    miss_val = df_with_placeholder.isnull().sum()
    miss_val_percent = 100 * df_with_placeholder.isnull().sum() / len(df)

    # Create the summary table
    miss_val_table = pd.concat([miss_val, miss_val_percent], axis=1)
    miss_val_table_ren_columns = miss_val_table.rename(columns = {
        0 : 'Missing values',
        1 : 'Missing % of total values'
        })

    miss_val_table_ren_columns = miss_val_table_ren_columns[
        miss_val_table_ren_columns.iloc[:,1] != 0
        ].sort_values('Missing % of total values', ascending=False).round(2)

    return miss_val_table_ren_columns

print("Missing values in train set:")
print(missing_values_table(df_num))
```

As it can be seen we have about ~19% of -200 values as placeholders!

```python
# Importing the imputers that will be used for different features

from sklearn.impute import KNNImputer
```

```python
# Spliting features from my dataset based on % of missing data from my function  low(<5%), medium(<30%), high(>30%)
low_missing_features = ['AH', 'RH', 'T', 'PT08.S5(O3)', 'PT08.S4(NO2)', 'PT08.S3(NOx)', 'PT08.S2(NMHC)', 'C6H6(GT)', 'PT08.S1(CO)']
medium_missing_features = ['NOx(GT)', 'NO2(GT)', 'CO(GT)']
high_missing_features = ['NMHC(GT)']

# Giving new reference to put modified dataset since i wont to keep df clean from anymore changes due to easier approach to clean the mistake if i make.
df_clean_1 = df.replace(-200, np.nan)

# Imputing all collumns for low % of missing data using meadian.
for feature in low_missing_features:
    df_clean_1[feature] = df_clean_1[feature].fillna(df_clean_1[feature].median())

# Imputing all collumns for medium % of missing data using KNN.
knn_imputer = KNNImputer(n_neighbors=5)
df_clean_1[medium_missing_features] = knn_imputer.fit_transform(df_clean_1[medium_missing_features])

for feature in high_missing_features:
    missing_percentage = df_clean_1[feature].isna().mean() * 100
    if missing_percentage > 30:
        df_clean_1.drop(columns=[feature], inplace=True)
```

### 2.2 Data Scailling

```python
# Importing libraries for Scailling
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import FunctionTransformer
```

```python
# Now first i will separate feature based on ther % difference between mean and median values from EDA-Univariate Analysis
# <10% Gaussian Like, >10% skewed & time feature

gaussian_features = ['NO2(GT)', 'PT08.S1(CO)', 'PT08.S3(NOx)','PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'AH', 'RH'] # PT08.S2(NMHC) is removed!
skewed_features = ['CO(GT)', 'C6H6(GT)', 'NOx(GT)'] # During research and learning i found that target feature is not required to be scaled but since data is skewed it can help with model robustness
time_features = ['Hour', 'DayOfWeek', 'Month'] # I will  not use year since it is not cyclical.

# Scailling for Gausian like data:
scaler = StandardScaler()
df_clean_1[gaussian_features] = scaler.fit_transform(df_clean_1[gaussian_features])

# Scailling for skewed data - log transformation & normalization
log_transformer = FunctionTransformer(np.log1p, validate=True)
df_clean_1[skewed_features] = log_transformer.fit_transform(df_clean_1[skewed_features])

min_max_scaler = MinMaxScaler()
df_clean_1[skewed_features] = min_max_scaler.fit_transform(df_clean_1[skewed_features])

# And here i will also transfrom my time data, explanation will be given in cell bellow as comment:

# Hour
df_clean_1['hour_sin'] = np.sin(2 * np.pi * df_clean_1['Hour'] / 24)
df_clean_1['hour_cos'] = np.cos(2 * np.pi * df_clean_1['Hour'] / 24)

# DayOfWeek
df_clean_1['dayofweek_sin'] = np.sin(2 * np.pi * df_clean_1['DayOfWeek'] / 7)
df_clean_1['dayofweek_cos'] = np.cos(2 * np.pi * df_clean_1['DayOfWeek'] / 7)

# Month
df_clean_1['month_sin'] = np.sin(2 * np.pi * df_clean_1['Month'] / 12)
df_clean_1['month_cos'] = np.cos(2 * np.pi * df_clean_1['Month'] / 12)

# Replacing transformed time data in df_clean_1 dataset
df_clean_1 = df_clean_1.drop(columns=['Hour', 'DayOfWeek', 'Month'])
```

About why i chose to transform time data.
* In one year we have 12 month, in one month 30 to 31 days and in day we have 24h. All those data repeat! And since 00 is close to 23h model can treat them as far apart. Same goes to days 1(mon) and 7(sun).
* So the reason to put those data into sin and cos should help model to be  more robust when dealing to time data.
* Ofcourse the reason why i needed sin and cos together is  due to mathematical nature of them both, sin gives us phase while cos gives us complementary angle and both of them gives us full periodic relation.

## <a name="inspection"> 3.0 Linear Regression without regularization

```python
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import KFold

from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as rmse
from sklearn.metrics import r2_score as r2
```

```python
# Predictors
X1 = sm.add_constant(df_clean_1.drop(columns=['C6H6(GT)', 'NOx(GT)', 'RH', 'Year', 'dayofweek_sin', 'dayofweek_cos', 'month_sin', 'month_cos']))
# Target
y1 = df_clean_1['C6H6(GT)']

X1, X1_test, y1, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=34)
X1_train, X1_valid, y1_train, y1_valid = train_test_split(X1, y1, test_size=.25, random_state=10)

# First to initialize the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X1_train, y1_train)

# Make predictions on the test set
y1_pred = model.predict(X1_test)

# To check the shape it needs to line up!

print(X1_valid.shape)
print(y1_valid.shape)

print(X1_train.shape)
print(X1_test.shape)
print(y1_train.shape)
print(y1_test.shape)
```

```python
print(f'Mean Absolute Error: {mae(y1_test, y1_pred)}')
print(f'R-squared Score: {r2(y1_test, y1_pred)}')
```

```python
model = sm.OLS(y1_train, X1_train).fit()
model.summary()
```

```python
# Used same code from workshop:
cv = KFold(n_splits=5, shuffle=True, random_state=2023)

# Create a list to store validation scores for each fold
r2_list = []
mae_list = []
rmse_list = []

# Loop through each fold in X and y
for train_ind, val_ind in cv.split(X1, y1):
    # Subset data based on CV folds
    X1_train, y1_train = X1.iloc[train_ind], y1.iloc[train_ind]
    X1_val, y1_val = X1.iloc[val_ind], y1.iloc[val_ind]
    # Fit the Model on fold's training data
    model = sm.OLS(y1_train, X1_train).fit()
    # Append Validation score to list
    r2_list.append(r2(y1_val, model.predict(X1_val)))
    mae_list.append(mae(y1_val, model.predict(X1_val)))
    rmse_list.append(rmse(y1_val, model.predict(X1_val)))

print("All Validation R2s: ", [round(x, 3) for x in r2_list])
print(f"Cross Val R2s: {round(np.mean(r2_list), 3)}")

print("All Validation MAEs: ", [round(x, 3) for x in mae_list])
print(f"Cross Val MAEs: {round(np.mean(mae_list), 3)}")

print("All Validation RMSEs: ", [round(x, 3) for x in rmse_list])
print(f"Cross Val RMSEs: {round(np.mean(rmse_list), 3)}")
```

Now to see residuals:

```python
from sklearn.linear_model import RidgeCV

n_alphas = 150
alphas = 10 ** np.linspace(-3, 3, n_alphas)

ridge_model = RidgeCV(alphas=alphas, cv=10)

ridge_model.fit(X1, y1)
print(f'R-squared Ridge Score: {ridge_model.score(X1, y1)}')
print(f'Ridge Mean Absolute Error: {mae(y1, ridge_model.predict(X1))}')
print(f'Optimal Ridge Alpha value: {ridge_model.alpha_}')
```

n_alphas - represent numbers of alphas we want to use. The more we have the more detailed search is but the computational time is slower and vice versa.

first 2 place under linspace() are representatin scale for our logaritmic scale
